```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import signal

pd.set_option('display.float_format', lambda x: '%e' % x)
```

```python
sim = pd.read_csv('sim1 - simulation_L_1.csv')
fig, ax = plt.subplots()
sim['sum'] = sim['v(b0)']*2**6 + sim['v(b1)']*2**5 + sim['v(b2)']*2**4 + sim['v(b3)']*2**3 + sim['v(b4)']*2**2 + sim['v(b5)']*2**1 + sim['v(b6)']*2**0
sim.plot(x='sum', y='i(viout)', ax = ax, marker='o', fillstyle = 'none', title = 'DAC Simulation Results')
plt.xlabel('bits sum')
plt.ylabel('Iout (A)')
plt.savefig("VtoI_plot.svg", format="svg")
```

```python
v_i = pd.read_csv('sim1 - simulation_L_1.csv')
fig, ax = plt.subplots()
v_i.plot(x='v-sweep', y='i(Vmeas)', ax = ax, title = 'Voltage To Current Characteristics')
plt.xlabel('Vin (V)')
plt.ylabel('Iout (A)')
plt.savefig("VtoI_plot.svg", format="svg")
error = ((v_i['i(Vmeas)'][0] - v_i['i(Vmeas)'][144])/v_i['i(Vmeas)'][0])/1.44
print(error)
```
